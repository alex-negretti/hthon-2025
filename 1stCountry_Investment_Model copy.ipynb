### Model parameters

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
```

```python
# --- 2. Define features and pre-processing steps ---
print("\n2. Defining feature types and pre-processing pipelines...")

# Identify numerical and categorical features
numerical_features = ['corona100_avg', 'contracts exclusivity_avg',
       'contracts experiential_avg', 'experiential_avg', 'influencers_avg',
       'national clients_avg', 'brandedcontent_avg', 'cinema_avg',
       'digitaldisplayandsearch_avg', 'digitalvideo_avg', 'facebook_avg',
       'instagram_avg', 'merchandiser_avg', 'ooh_avg', 'opentv_avg',
       'paytv_avg', 'pinterest_avg', 'print_avg', 'radio_avg',
       'socialmedia_avg', 'trade_off_avg', 'trade_on_avg', 'twitter_avg',
       'youtube_avg', 'corona100_std', 'contracts exclusivity_std',
       'contracts experiential_std', 'experiential_std', 'influencers_std',
       'national clients_std', 'brandedcontent_std', 'cinema_std',
       'digitaldisplayandsearch_std', 'digitalvideo_std', 'facebook_std',
       'instagram_std', 'merchandiser_std', 'ooh_std', 'opentv_std',
       'paytv_std', 'pinterest_std', 'print_std', 'radio_std',
       'socialmedia_std', 'trade_off_std', 'trade_on_std', 'twitter_std',
       'youtube_std', 'corona100_sum', 'contracts exclusivity_sum',
       'contracts experiential_sum', 'experiential_sum', 'influencers_sum',
       'national clients_sum', 'brandedcontent_sum', 'cinema_sum',
       'digitaldisplayandsearch_sum', 'digitalvideo_sum', 'facebook_sum',
       'instagram_sum', 'merchandiser_sum', 'ooh_sum', 'opentv_sum',
       'paytv_sum', 'pinterest_sum', 'print_sum', 'radio_sum',
       'socialmedia_sum', 'trade_off_sum', 'trade_on_sum', 'twitter_sum',
       'youtube_sum', 'meaning', 'difference',
       'salience', 'power', 'premium']
categorical_features = ['quarter', 'brand']

# Create a preprocessor using ColumnTransformer
# It will apply different transformations to different columns in parallel.
# OneHotEncoder is used for categorical features.
# StandardScaler is used for numerical features.
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])

```

```python
y = whole_date['power_lag1'].fillna(0)
```

```python
# --- 3. Split data and train the XGBoost model ---
print("\n3. Splitting data and training the XGBoost model...")

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(whole_date, y, test_size=0.2, random_state=42)

# Create a pipeline to combine preprocessing and the model
# This ensures that all preprocessing is applied correctly to both training and test data.
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, random_state=42))
])

# Train the model
model_pipeline.fit(X_train, y_train)

print("\nXGBoost model training complete.")

# --- 4. Evaluate the model ---
print("\n4. Evaluating the model on the test set...")

# Make predictions on the test set
y_pred = model_pipeline.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"\nModel Mean Squared Error on test set: {mse:.4f}")
print(f"Model Root Mean Squared Error on test set: {rmse:.4f}")
```

```python
import matplotlib.pyplot as plt

# --- 5. Plot the actual vs. predicted values ---
print("\n5. Plotting actual vs. predicted values...")
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.6, color='dodgerblue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred)')
plt.title('Actual vs. Predicted Values')
plt.grid(True)
plt.show()
```

```python
feature_names_out = model_pipeline.named_steps['preprocessor'].get_feature_names_out()

# Get the fitted XGBoost regressor
regressor = model_pipeline.named_steps['regressor']

# Get the feature importances
importance = regressor.feature_importances_

# Create a DataFrame for better visualization
feature_importance_df = pd.DataFrame({
    'feature': feature_names_out,
    'importance': importance
}).sort_values(by='importance', ascending=False)
```

```python
# Plot the feature importance
plt.figure(figsize=(12, 15))
plt.barh(feature_importance_df['feature'], feature_importance_df['importance'], color='teal')
plt.xlabel('Feature Importance (Gain)')
plt.ylabel('Feature')
plt.title('Feature Importance from XGBoost Model')
plt.gca().invert_yaxis()  # Invert y-axis to show the most important feature at the top
plt.show()
```

```python
# Select the top 15 most important features
# The number of features in this synthetic dataset is small, so we will select all of them.
# In a real-world scenario, you would use a value like `top_n = 15`.
top_n = len(feature_importance_df)
top_features = feature_importance_df.head(top_n)['feature'].tolist()

# Get the preprocessor from the original pipeline
preprocessor_fitted = model_pipeline.named_steps['preprocessor']

# Transform the original data using the fitted preprocessor
X_train_transformed = preprocessor_fitted.transform(X_train)
X_test_transformed = preprocessor_fitted.transform(X_test)

# Create a DataFrame from the transformed data to easily select columns by name
# We need to map the output of the transformer back to the feature names
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=feature_names_out)
X_test_transformed_df = pd.DataFrame(X_test_transformed, columns=feature_names_out)

# Filter the transformed data to only include the top features
X_train_filtered = X_train_transformed_df[top_features]
X_test_filtered = X_test_transformed_df[top_features]
```

```python
from sklearn.model_selection import GridSearchCV

print("\n7. Performing Hyperparameter Tuning with GridSearchCV...")

# Define the parameter grid to search
# Note: When using a pipeline, the parameter names must be prefixed with the step name (e.g., 'regressor__').
param_grid = {
    'regressor__n_estimators': [50, 100, 200, 500, 1000],
    'regressor__learning_rate': [0.01, 0.1, 0.2, 0.3],
    'regressor__max_depth': [3, 5, 7, 10],
    'regressor__subsample': [0.7, 0.9, 1.0]
}

# Initialize the GridSearchCV object
# 'neg_root_mean_squared_error' is used because GridSearchCV maximizes the score,
# so we use the negative RMSE to find the minimum error.
grid_search = GridSearchCV(
    estimator=xgb.XGBRegressor(objective='reg:squarederror', random_state=42),
    param_grid=param_grid,
    scoring='neg_root_mean_squared_error',
    cv=5,
    n_jobs=-1,  # Use all available cores
    verbose=2
)

# Fit the grid search to the data
grid_search.fit(X_train_filtered, y_train)

# Print the best parameters and the best score
print("\nGrid Search completed.")
print(f"Best hyperparameters found: {grid_search.best_params_}")
best_rmse = -grid_search.best_score_
print(f"Best cross-validation RMSE: {best_rmse:.4f}")

# Optionally, you can now use the best estimator to make predictions
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)
tuned_rmse = np.sqrt(mean_squared_error(y_test, y_pred_tuned))
print(f"Tuned model's RMSE on test set: {tuned_rmse:.4f}")
```

```python
import matplotlib.pyplot as plt

# --- 5. Plot the actual vs. predicted values ---
print("\n5. Plotting actual vs. predicted values...")
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_tuned, alpha=0.6, color='dodgerblue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual Values (y_test)')
plt.ylabel('Predicted Values (y_pred_tuned)')
plt.title('Actual vs. Predicted Values')
plt.grid(True)
plt.show()
```
